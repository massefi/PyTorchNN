```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

scaler = StandardScaler()
X_train_sclaed = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_sclaed
```

```python
y_train
```

```python
X_train_sclaed_tensor = torch.from_numpy(X_train_sclaed).float()
X_test_scaled_tensor = torch.from_numpy(X_test_scaled).float()
y_train_tensor = torch.from_numpy(y_train).float().unsqueeze(1)
y_test_tensor = torch.from_numpy(y_test).float().unsqueeze(1)
```

```python
train_dataset = TensorDataset(X_train_sclaed_tensor, y_train_tensor)
```

```python
X_train_sclaed_tensor.shape
```

```python
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
```

```python
class BCNet(nn.Module):
    def __init__(self):
        super(BCNet, self).__init__()
        self.fcl1 = nn.Linear(30, 64)

        self.fc2 = nn.Linear(64, 32)
        self.fc3 = nn.Linear(32, 1)
        self.dropout = nn.Dropout(0.5)
        # self.relu = nn.ReLU()
        # self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = F.relu(self.fcl1(x))
        x = F.relu(self.fc2(x))
        X = self.dropout(x)
        x = F.sigmoid(self.fc3(x))

        return x
```

```python
model = BCNet()
```

```python
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
epochs = 20

for epoc in range(epochs):
  model.train()
  running_loss = 0.0

  for x_batch, y_batch in train_loader:
    optimizer.zero_grad()
    preds = model(x_batch)
    loss = criterion(preds, y_batch)
    loss.backward()
    optimizer.step()
    running_loss += loss.item()
  print(f'Epoch: {epoc+1}, Loss: {running_loss/len(train_loader)}')
```

```python
with torch.no_grad():
  model.eval()

  preds = model(X_test_scaled_tensor)
  # preds = (preds > 0.5).float
  loss = criterion(preds, y_test_tensor).item()

  accuracy = ((preds > 0.5).float() == y_test_tensor).float().mean().item()
```
